```python
#How many distinct (different) actors' last names are there?
query1 = '''select rental_id, rental_date, inventory_id, customer_id, return_date, staff_id, last_update, 
CASE
when weekday(rental_date) = 0 OR  1 OR 2 OR 3 OR 4 then 'weekday'
when weekday(rental_date) = 5 OR 6 then 'weekend'
END AS 'day_type'
from sakila.rental;'''

weekday = pd.read_sql_query(query1, engine)
weekday
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

import datetime
day_type = []
for i in rental["rental_date"]:
    if i.weekday() < 5:
        day_type.append("Weekday")
    elif i.weekday() >= 5:
        day_type.append("Weekend")
rental["day_type"] = day_type 
display(rental)
```
